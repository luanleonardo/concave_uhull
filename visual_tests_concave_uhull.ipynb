```python
from concave_uhull import alpha_shape_polygons
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
x = 4 * np.random.rand(5000)
y = 4 * np.random.rand(5000)
coordinates_square_set = list(zip(x, y))
plt.plot(x, y, ".")
```

```python
polygons = alpha_shape_polygons(coordinates_square_set)
len(polygons)
```

```python
plt.figure()
plt.axis("equal")
x, y = zip(*coordinates_square_set)
plt.plot(x, y, ".")
for poly in polygons:
    x, y = zip(*poly)
    plt.plot(x, y)
plt.show()
```

```python
from concave_uhull import euclidean_distance

# centro dos circulos
center_circles = (2.0, 2.0)

# checa se ponto esta na coroa circular
def _is_circular_crown_point(point, center):
    return 1.0 < euclidean_distance(point, center) < np.sqrt(2.0)


# retorna pontos na coroa circular
circular_crown_set = list(
    filter(
        lambda point: _is_circular_crown_point(point, center_circles),
        coordinates_square_set,
    )
)
x, y = zip(*circular_crown_set)
plt.plot(x, y, ".")
```

```python
polygons = alpha_shape_polygons(circular_crown_set)
len(polygons)
```

```python
plt.figure()
plt.axis("equal")
x, y = zip(*circular_crown_set)
plt.plot(x, y, ".")
for poly in polygons:
    x, y = zip(*poly)
    plt.plot(x, y)
plt.show()
```

## Points in Par, Brazil

```python
import json

# loggibud dataset download: https://loggibud.s3.amazonaws.com/dataset.zip
with open(
    "dataset/data/delivery-instances-1.0/train/pa/pa-0.json"
) as dataset_file:
    data = json.load(dataset_file)
```

```python
coordinates_points = [
    (delivery["point"]["lng"], delivery["point"]["lat"])
    for delivery in data["deliveries"]
]
```

```python
polygons = alpha_shape_polygons(coordinates_points, alpha=10)

plt.figure()
plt.axis("equal")
x, y = zip(*coordinates_points)
plt.plot(x, y, ".")
for poly in polygons:
    x, y = zip(*poly)
    plt.plot(x, y)
plt.show()
```
